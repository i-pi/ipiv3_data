```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

plt.rcParams['font.size'] = 15
plt.rc('axes',linewidth=2,labelpad=10)
plt.rcParams["xtick.direction"] = "in"
plt.rcParams["ytick.direction"] = "in"
plt.rc('xtick.major',size=10, width=2)
plt.rc('xtick.minor',size=7, width=2)
plt.rc('ytick.major',size=10, width=2)
plt.rc('ytick.minor',size=7, width=2)

label = 'longest'
x = np.loadtxt('data/%s_x' % label)
y = np.loadtxt('data/%s_y' % label)
err = np.loadtxt('data/%s_err' % label)
plt.errorbar(x, y, err, linestyle='None', marker='o', label=label)

plt.axhline(1.0, linestyle='--')

plt.xlabel(r'$\beta \hbar \omega_0$')
plt.ylabel(r'exchange-all $/$ $(1/N)$')
# plt.legend()

plt.savefig('bosonic_gauge_16_3_beads_5000000.pdf', bbox_inches='tight')
plt.show()
```
