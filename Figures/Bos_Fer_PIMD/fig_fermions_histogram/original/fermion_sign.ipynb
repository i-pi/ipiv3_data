```python
import numpy as np
from matplotlib import pyplot as plt
import re

plt.rcParams['font.size'] = 15
plt.rc('axes',linewidth=2,labelpad=10)
plt.rcParams["xtick.direction"] = "in"
plt.rcParams["ytick.direction"] = "in"
plt.rc('xtick.major',size=10, width=2)
plt.rc('xtick.minor',size=7, width=2)
plt.rc('ytick.major',size=10, width=2)
plt.rc('ytick.minor',size=7, width=2)


def plot_est_in_time_for_temp(temperature_kelvin):
    bin_edges = np.loadtxt('data/%s_bosonic_hist_edges.txt' % str(temperature_kelvin))
    density = np.loadtxt('data/%s_bosonic_hist_density.txt' % str(temperature_kelvin))
    plt.plot(bin_edges[:-1], density, 'o', label='bosonic energy')
    bin_edges = np.loadtxt('data/%s_fermionic_hist_edges.txt' % str(temperature_kelvin))
    density = np.loadtxt('data/%s_fermionic_hist_density.txt' % str(temperature_kelvin))
    plt.plot(bin_edges[:-1], density, 'o', label='fermionic energy')

    plt.xlabel(r'$\langle E \rangle / \hbar \omega_0$')
    plt.ylabel('hist')
    plt.legend()

    plt.savefig('ipi_noninteracting-fermions_energy_hist_3_%s.pdf' % str(temperature_kelvin), bbox_inches='tight')
    plt.show()
```

```python
plot_est_in_time_for_temp(23.22)
```

```python
plot_est_in_time_for_temp(46.46)
```
